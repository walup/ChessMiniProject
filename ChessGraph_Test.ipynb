```python
from ChessGame import CoordinateTranslator
from ChessGame import Chessboard
from ChessGame import ChessMove
from ChessGame import ChessPieceColor
from ChessGraph import Chessgraph
import matplotlib.pyplot as plt
```

## Graph degree distribution test

```python
board = Chessboard()
#E4
board.makeMove(ChessMove(["E",2], ["E",4], False), False)
#D4
board.makeMove(ChessMove(["E",7], ["E",5], False), False)
#Knight to F3
board.makeMove(ChessMove(["G",1], ["F",3], False), False)
#Knight to C6
board.makeMove(ChessMove(["B", 8], ["C", 6], False), False)
#Bishop to C4
board.makeMove(ChessMove(["F", 1], ["C", 4], False), False)
#Bishop to C5
board.makeMove(ChessMove(["F",8], ["C",5], False), False)
board.displayBoard()
```

```python
chessGraph = Chessgraph()
chessGraph.initializeFromChessboard(board)
chessGraph.displayGraph()
```

```python
degreeValues, degreeDistribution = chessGraph.getDegreeDistribution()
print(degreeDistribution[0])
print(degreeDistribution)
```

```python
plt.figure()
plt.bar(degreeValues, degreeDistribution, facecolor = "#24a4ff")
plt.xlabel("Degrees")
plt.ylabel("Frequency")
```
